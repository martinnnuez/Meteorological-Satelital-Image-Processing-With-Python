Creamos una tupla y una lista iguales y las imprimimos en pantalla. Asi mismo accedemos al *help* de cada una. Que diferencias se observan?

### Funciones y sus gráficos. Loops y operadores lógicos

A continuación se define una función a trozos (con **def**) para lo cual se utiliza la función **if**. Además usamos loop (**for**) y vemos como añadir elementos al array. Para los bloques ed funciones observar la importancia de la identación.

Tarea: Ejecutar separando en bloques, probar modificaciones y observar los resultados

```python
#%%
from matplotlib import pyplot as plt #se llama a una sublibrería específica
print("\n Creo una funcion linal a trozos")

def lineal_a_trozos(x):#devuelve el valor de la funcion lineal a trozos
  '''Esta función devuelve el valor de una función lineal a trozos para un valor 
  de x entrado. Los intervalos lineales están determinados dentro de la función
  '''
  xx = np.array([0, 9, 24]) #valores de x que definen los segmentos 
  yy = np.array([2, 5, 3]) #valores de y que definen los segmentos
  if x < xx[0] or x > xx[2]:
    print("x=%f no pertence al intervalo [%.2f,%.3f]" %(x,xx[0],xx[2])) #print con formato
    return
  elif x <= xx[1]: # else if
    f = (x - xx[0])/(xx[1]-xx[0])*(yy[1]-yy[0])+yy[0]
  elif x <= xx[2]: # else if
    f = (x - xx[1]) / (xx[2] - xx[1]) * (yy[2] - yy[1]) + yy[1]
  return f

print("\nConstruyo un vector para los valores del intervalo con dx = 1.5")

x = np.linspace(0, 24, 20)
f = np.array([])#creo un vector vacio

for i in x: #el loop llama a los elementos del vector sin necedidad de indexar
    f = np.append(f, lineal_a_trozos(i))#añade al vector f un elemento

print ("\nGráfico la funcion")
       
plt.figure()
plt.plot(x, f, '.g')
plt.title("x vs f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
#%%
print("\nCreo una funcion lineal a trozos generalizada")

def lineal_a_trozos2(x, xx, yy):#devuelve el valor de la funcion para a<=x<=b
  '''Esta función devuelve el valor un array de una función lineal a trozos para 
  un array ingresado.
  Esta función es completamente flexible pues los intervalos son definidos a partir
  de los pares (xx, yy) ingresados como paramétros a la función. 
  '''
  if x < xx[0] or x > xx[-1]:
    print("x=%f no pertence al intervalo [%.2f,%.3f]" %(x, xx[0], xx[2])) #print con formato
    return #sale de la funcion
  elif x == xx[-1]:
    f = yy[-1]
  else:
    for i in range(xx.size-1):
      if x >= xx[i] and x < xx[i+1]: # else if
        f=(x - xx[i]) / (xx[i+1] - xx[i]) * (yy[i+1] - yy[i]) + yy[i]
  return f

x = np.linspace(0, 100, 200)
f = np.zeros(x.size)#creo un vector de dimension igual a x con zeros

xx = np.array([0, 11, 14, 40, 100],dtype=float)
yy = np.array([300, 210, 210, 270, 200])

for i in range(x.size): #el loop llama a los elementos del vector sin necedidad de indexar
    f[i] = lineal_a_trozos2(x[i], xx, yy)

print ("\nGráfico la funcion")
       
plt.figure()
plt.plot(x, f, '*b')
plt.title("x vs f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

print ("\nGráfico la funcion en escala logaritmica para x")
       
plt.figure()
plt.semilogx(x, f, 'r')
plt.title("x vs f(x), x en escala logaritmica")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()      
```
